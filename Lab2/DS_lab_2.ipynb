```python
#все вище - спроби очистити дані
#на жаль цьому нас не вчили PoroSad  

#апдейт від 14.11, я попитав у однокласника, що навчається в КПІ, він також не хнає, 
#що робити з текстом, що сприймається як object
#-----------------------------------------------------------------
data = pd.read_excel("rollingsales_manhattan.xls",parse_dates = True, verbose=True, na_values=[' ','  ','   ','\t','\n','\n\n',' \n'])#тому буду чистити частково, наскільки зможу
data
```

```python
data.isnull().sum()
```

```python
#невід'ємна ціна
cleared_data = data[data["SALE PRICE"]>0]
#виключаю без району, умова виглядає як костиль
cleared_data = cleared_data[cleared_data["NEIGHBORHOOD"].str.isspace() != True]
#видаляю непотрібні стовпці
cleared_data = cleared_data.drop(columns=["EASE-MENT","LOT","BLOCK","APARTMENT NUMBER","ZIP CODE"])
cleared_data
```

```python
cleared_data.groupby(cleared_data["SALE DATE"].dt.year).sum().plot(kind="pie", y="SALE PRICE", autopct='%.1f', legend='')
```

```python
cleared_data.groupby([cleared_data["SALE DATE"].dt.year, cleared_data["SALE DATE"].dt.month]).sum()#10 нулів
```

```python
cleared_data.groupby([cleared_data["SALE DATE"].dt.year, cleared_data["SALE DATE"].dt.month]).sum().plot(
    kind="bar", y="SALE PRICE",legend="",xlabel="Sale date by month",ylabel="Sum sale price\n1:10^10$")
```

```python
cleared_data.groupby(cleared_data["NEIGHBORHOOD"]).sum()
```

```python
cleared_data.groupby(cleared_data["NEIGHBORHOOD"]).sum().plot(
    kind="bar", y="SALE PRICE",legend="",xlabel="Neighborhood",ylabel="Sum sale price\n1:10^9$", figsize=(10,5))
```

```python
def year_built_classify(number):
    if number <= 1800:
        return "<=1800"
    elif number <= 1850:
        return "1801-1850"
    elif number <= 1900:
        return "1851-1900"
    elif number <= 1950:
        return "1901-1950"
    elif number <= 2000:
        return "1951-2000"
    else:
        return "2000>"

grouped_data = cleared_data[cleared_data["YEAR BUILT"]>1000]
grouped_data["year_built_group"] = grouped_data["YEAR BUILT"].apply(lambda x:year_built_classify(x))
grouped_data
```

```python
grouped_data.groupby(grouped_data["year_built_group"]).count().plot(
    kind="bar", y="BOROUGH",legend="",xlabel="Construction year",ylabel="Number of buildings", figsize=(5,15))
```
